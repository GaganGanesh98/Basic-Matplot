```python
a = np.linspace(0,10,11) #uses NumPy's linspace function to create an array of evenly spaced numbers over a specified interval, start = 0, stop = 10, num of elements = 11.
```

```python
a #as you can see we got 11 elements ranging from 0 to 10.
```

```python
b = a**4
b
```

```python
#plt.figure() will create an empty plot.
fig = plt.figure(figsize = (7,8),dpi=200) #dpi will make the numbers look clearer on your plot index # figsize will resize the plot.
axis1 = fig.add_axes([0,0,1,1]) #add_axes:adds a set of axes to the figure([left=0, bottom=0, width=1, height=1])
axis1.plot(a,b) #Plots the data a on the x-axis and b on the y-axis.


#you can modify the table as required without disturbing the other plots in the diagram.
axis1.set_xlim(0,8) #set_xlim(0, 8) sets the x-axis limits from 0 to 8.
axis1.set_ylim(0,8000) #sets the y-axis limits from 0 to 8000.
axis1.set_xlabel('x1') #label x
axis1.set_ylabel('y1') #label y


axis2 = fig.add_axes([0.5,0.2,0.25,0.20]) #plot another graph into the graph
axis2.plot(a,b)

axis2.set_xlim(0,10)
axis2.set_ylim(0,10000)
axis2.set_xlabel('x2')
axis2.set_ylabel('y2')
axis2.set_title('zoom')

#now save with .savefig()
fig.savefig('MATPLOT2',bbox_inches='tight') #bbox-inches or boundingbox = tight will show the x and y axis with figures in the plot or else it will only show the plot without any figures.
```

how to create a grid of subplots using plt.subplots, customize them, and save the final figure.

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2) #creates a 2x2 grid of subplots, now it is a 2dimentional arrey 
axes[0][0].plot(x,y) #refers to the top-left subplot.
axes[1][1].plot(a,b) #refers to the bottom-right subplot.
plt.tight_layout() #makes sure no axes or figures are overlaping eachother and makes your plot look need by giving space between them.
fig.subplots_adjust(wspace=1,hspace=1) #or we can manually adjust the spacing betweed the axes using subplots_adjust(wspace= for width spacing , hspace= for hight spacing)

axes[1][1].set_title('LAST') #Sets a title for the bottom-right subplot.
axes[0][1].set_xlabel('gotcha') #Sets an x-axis label for the top-right subplot.


b = a**4
y = x*2
axes[1][0].plot(a,b)

fig.suptitle('DEMO',fontsize = 16) #Adds a title to the entire figure with a font size of 16.
fig.savefig('DEMO.png',bbox_inches = 'tight')
```

This code demonstrates how to create a plot with multiple lines, customize their appearance (color, line style, thickness, markers, etc.), and add them to a single set of axes:

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='blue',lw = 5) #Plots x vs x (a straight line), sets the line color to blue, sets the line width (thickness) to 5.
ax.plot(x,x+1,color='pink',ls = '--',marker = 'o', ms = 5) # (a straight line shifted upward by 1, sets the line color to pink, sets the line style to dashed, adds circular markers at each data point, sets the marker size to 5)
```

This code creates a 3D line plot using Matplotlib's mpl_toolkits.mplot3d module. It plots a 3D curve.

```python
import matplotlib.pyplot as plt         #is used for plotting.
from mpl_toolkits.mplot3d import Axes3D #enables 3D plotting.
import numpy as np                      #is used for numerical operations and generating data.

# Data for the plot
x = np.linspace(-5, 5, 100)
y = np.sin(x)
z = np.cos(x)

# Create a figure
fig = plt.figure()

# Add a 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D line
ax.plot3D(x, y, z, label='3D Line') #Plots a 3D line using x, y, and z. label='3D Line' assigns a label to the line for the legend.
ax.set_title('3D Line Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.legend() 
plt.show()
```

This code demonstrates how to plot Einstein's famous equation 
E = mc^2 

```python
m = np.linspace(0,10,11)
c = 3 * 10 ** 8 #speed of light
E = m*c**2      #E = mc^2 
```

```python
plt.plot(m,E,color = 'red', lw = 5) #mass is x axis, E is y axis.
plt.title('E=mc^2')
plt.xlabel('Mass')
plt.ylabel('Energy')
plt.xlim(0,10)
plt.show()
```
